#### Q1: Income dataset

```python
#importing pandas library for data analysis
import pandas as pd
```

```python
df = pd.read_csv('incomedata.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#Select the variables age category (agecat) and gender for the subjects 
# who did not complete high school

df[df['educ'] == 'Did not complete high school'][['agecat','gender']]
```

```python
#Select the variables age category and gender for the subjects 
# who did not complete high school and have the income greater than 20

df[(df['educ'] == 'Did not complete high school') & (df['income'] > 20)][['agecat','gender']]
```

```python
#Select the male subjects with the income lower than 65 and compute their average income

print('Average income of male subjects with income lower than 65: ',df[(df['gender'] == 'Male') & (df['income'] < 65)]['income'].mean())
```

```python
#Group the subjects by age category and compute the mean and the standard deviation of the income

df.groupby(by = 'agecat')['income'].agg(['mean', 'std'])
```

```python
#Sort the data frame by income in ascending order

df_sorted_asc = df.sort_values(by = 'income', ascending= True)
df_sorted_asc
```

```python
#Sort the data frame by income in descending order
df_sorted_desc = df.sort_values(by = 'income', ascending= False)
df_sorted_desc
```

```python
#Count the number of subjects in each education category

df['educ'].value_counts()
```
